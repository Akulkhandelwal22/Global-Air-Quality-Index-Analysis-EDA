Explanation of common AQI terms:
PM2.5: PM2.5 refers to tiny particles or droplets in the air that are 2.5 micrometers or less in width. They can be harmful to human health when inhaled, especially in high concentrations.

Ozone: Ozone is a gas that can form in the atmosphere through a chemical reaction between sunlight and other pollutants. High levels of ozone can be harmful to human health, particularly for those with respiratory issues.

Carbon Monoxide (CO): CO is a colorless, odorless gas that is produced by the incomplete burning of fossil fuels. High levels of CO can be toxic to humans and can cause headaches, dizziness, and nausea

Attribute Information:
1.Country: Name of the country.

2.City: Name of the city within the country.

3.AQI Value: Numeric value representing the Air Quality Index (AQI) for the given location.

4.AQI Category: Categorical value indicating the air quality category based on the AQI value.

5.CO AQI Value: Numeric value representing the Carbon Monoxide (CO) Air Quality Index (AQI) for the given location.

6.CO AQI Category: Categorical value indicating the CO AQI category based on the CO AQI value.

7.Ozone AQI Value: Numeric value representing the Ozone Air Quality Index (AQI) for the given location.

8.Ozone AQI Category: Categorical value indicating the Ozone AQI category based on the Ozone AQI value.

9.NO2 AQI Value: Numeric value representing the Nitrogen Dioxide (NO2) Air Quality Index (AQI) for the given location.

10.NO2 AQI Category: Categorical value indicating the NO2 AQI category based on the NO2 AQI value.

11.PM2.5 AQI Value: Numeric value representing the Particulate Matter 2.5 (PM2.5) Air Quality Index (AQI) for the given location.

12.PM2.5 AQI Category: Categorical value indicating the PM2.5 AQI category based on the PM2.5 AQI value.

13.lat: Latitude coordinate of the city.

14.lng: Longitude coordinate of the city.

```python
import pandas as pd
from matplotlib import pyplot as plt
data=pd.read_csv("D:\Python\AQI and Lat Long of Countries.csv")
data.info()
```

```python
#Count the number of country in the dataset

num_countries = data['Country'].nunique()
print("Number of countries: ", num_countries)
```

```python
#Find the average AQI Value for each country

average_aqi = data.groupby('Country')['AQI Value'].mean().reset_index()
print("Average AQI Value for each counrty:")
print(average_aqi)
```

```python
#Retrieve all rows where the AQI category is "Good" and the CO AQI Category is "Good"

filtered_data=data[(data['AQI Category']=='Good') & (data['CO AQI Category']=='Good')]
plt.scatter(filtered_data['lat'],filtered_data['lng'])
plt.xlabel('Latitiude')
plt.ylabel('Longitude')
plt.title('Locations With Good AQI category and Good CO AQI Category')
plt.show()
```

```python
#Calculate the median PM2.5 AQI Value for a specific country.

target_country = input('Enter Country: ')
country_data = data[data['Country'] == target_country]
median_pm25_aqi = country_data['PM2.5 AQI Value'].median()

plt.hist(country_data['PM2.5 AQI Value'], bins=10)
plt.xlabel('PM2.5 AQI Value')
plt.ylabel('Frequency')
plt.title(f'PM2.5 AQI Values for {target_country}')
plt.axvline(x=median_pm25_aqi, color='red', linestyle='--', linewidth=2, label='Median')
plt.legend()
plt.show()
print("Median PM2.5 AQI Value for a specific country",median_pm25_aqi)
```

```python
#Get the top 5 cities with the highest NO2 AQI Value

sorted_data = data.sort_values('NO2 AQI Value', ascending=False)
top_cities = sorted_data.head(5)['City']

plt.bar(top_cities, sorted_data.head(5)['NO2 AQI Value'])
plt.xlabel('City')
plt.ylabel('NO2 AQI Value')
plt.title('Top 5 Cities with Highest NO2 AQI Value')
plt.xticks(rotation=90)
plt.show()
print("top 5 cities with the highest NO2 AQI Value are:")
print(top_cities)
```

```python
#Count the number of rows where the Ozone AQI Value is above a certain threshold.

threshold = 100
filtered_data = data[data['Ozone AQI Value'] > threshold]
count = len(filtered_data)
print("number of rows where the Ozone AQI Value is above a certain threshold: ",count)
```

```python
#Retrieve all rows where the AQI Category is "Hazardous" or the PM2.5 AQI Category is "Unhealthy

filtered_data = data[(data['AQI Category'] == 'Hazardous') | (data['PM2.5 AQI Category'] == 'Unhealthy')]

plt.scatter(filtered_data['lat'], filtered_data['lng'])
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Locations with Hazardous AQI Category or Unhealthy PM2.5 AQI Category')
plt.show()
```

```python
#Find the city with the highest AQI Value in each country

max_aqi_cities = data.groupby('Country')['AQI Value'].idxmax()
cities_data = data.loc[max_aqi_cities, ['Country', 'City', 'AQI Value']]
print(cities_data)
```

```python
#Calculate the percentage of cities with "Unhealthy" air quality

unhealthy_cities = data[data['AQI Category'] == 'Unhealthy']
total_cities = len(data)
percentage = (len(unhealthy_cities) / total_cities) * 100

labels = ['Unhealthy', 'Healthy']
sizes = [percentage, 100 - percentage]
plt.pie(sizes, labels=labels,autopct='%1.1f%%')
plt.axis('equal')
plt.title('Percentage of Cities with Unhealthy Air Quality')
plt.show()
print("The percentage of unhealthy cities is ",percentage)
```

```python
#Count the number of cities with a latitude value between two specified values.

lower_latitude = 50
upper_latitude = 100
filtered_data = data[(data['lat'] >= lower_latitude) & (data['lat'] <= upper_latitude)]
city_count = len(filtered_data)
print("Number of cities with a latitude value between two specified values ",city_count)
```

```python
#Find the country with the highest number of cities in the "Unhealthy" AQI Category.

unhealthy_cities = data[data['AQI Category'] == 'Unhealthy']
country_counts = unhealthy_cities['Country'].value_counts()
country_highest = country_counts.idxmax()
print("Country with the highest number of cities in the Unhealthy AQI Category ",country_highest)
```

```python
#Calculate the standard deviation of the AQI Value for each country.

std_deviation = data.groupby('Country')['AQI Value'].std()
print("Standard deviation of the AQI Value for each country: ")
print(std_deviation)
```

```python
#Get the latitude and longitude of the city with the highest PM2.5 AQI Value in a specific country

country = input('Enter Country: ')
country_data = data[data['Country'] == country]
highest_pm25_city = country_data.loc[country_data['PM2.5 AQI Value'].idxmax()]
latitude = highest_pm25_city['lat']
longitude = highest_pm25_city['lng']

print("Latitude of the city with the highest PM2.5 AQI Value in a specific country",latitude)
print("Longitude of the city with the highest PM2.5 AQI Value in a specific country",longitude)
```

```python
#Retrieve all rows where the AQI Value is equal to the minimum AQI Value in the dataset

min_aqi_value = data['AQI Value'].min()
min_aqi_rows = data[data['AQI Value'] == min_aqi_value]
print(min_aqi_rows)
```

```python
#Count the number of rows where the Ozone AQI Category is "Good" and the NO2 AQI Category is "Moderate".

filtered_data = data[(data['Ozone AQI Category'] == 'Good') & (data['NO2 AQI Category'] == 'Moderate')]
count = len(filtered_data)
print("Number of rows where the Ozone AQI Category is Good and the NO2 AQI Category is Moderate ",count)
```

```python
#Calculate the range (maximum - minimum) of the CO AQI Value for each country.

co_range = data.groupby('Country')['CO AQI Value'].apply(lambda x: x.max() - x.min())
print("the range (maximum - minimum) of the CO AQI Value for each country",co_range)
```

```python
#Find the city with the lowest AQI Value in each country

min_aqi_cities = data.groupby('Country')['AQI Value'].idxmin()
cities_data = data.loc[min_aqi_cities, ['Country', 'City', 'AQI Value']]
print(cities_data)
```

```python
#Retrieve all rows where the AQI Category is "Hazardous" and the Ozone AQI Category is "Unhealthy".

filtered_data = data[(data['AQI Category'] == 'Hazardous') & (data['Ozone AQI Category'] == 'Unhealthy')]
print("rows where the AQI Category is Hazardous and the Ozone AQI Category is Unhealthy :")
print(filtered_data)
```

```python
#Get the top 5 countries with the highest percentage of cities in the "Good" AQI Category.

good_aqi_cities = data[data['AQI Category'] == 'Good']
country_counts = good_aqi_cities['Country'].value_counts()
country_percentages = (country_counts / data['Country'].value_counts()) * 100
top_countries = country_percentages.nlargest(5)

plt.bar(top_countries.index, top_countries.values)
plt.xlabel('Country')
plt.ylabel('Percentage of Cities in "Good" AQI Category (%)')
plt.title('Top 5 Countries with Highest Percentage of Cities in "Good" AQI Category')
plt.xticks(rotation=45)
plt.show()
print("Top 5 countries with the highest percentage of cities in the Good AQI Category :")
print(top_countries)
```

```python
#Count the number of rows where the  NO2 AQI Value is below a given threshold and the PM2.5 AQI Value is above another threshold

pm25_threshold = 50
no2_threshold = 10
filtered_data = data[(data['PM2.5 AQI Value'] < pm25_threshold) & (data['NO2 AQI Value'] > no2_threshold)]
num_rows = len(filtered_data)
print("Number of rows where the NO2 AQI Value is below a given threshold and the PM2.5 AQI Value is above another threshold: "
      ,num_rows)
```

```python
#Find the country with the highest overall air quality (lowest average AQI Value) in a specific city

city_name=input("Enter city: ")
city_df = data[data['City'] == city_name]
average_aqi_by_country = city_df.groupby('Country')['AQI Value'].mean()
country_with_lowest_aqi = average_aqi_by_country.idxmin()
print("Country with the highest overall air quality in", city_name, country_with_lowest_aqi)
```

```python
#Retrieve all rows where the AQI Category is "Unhealthy" and the NO2 AQI Category is "Good".

filtered_data = data[(data['AQI Category'] == 'Unhealthy') & (data['NO2 AQI Category'] == 'Good')]

plt.scatter(filtered_data['lat'], filtered_data['lng'], alpha=0.7)
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Locations: AQI Category = "Unhealthy" and NO2 AQI Category = "Good"')
plt.show()
```

```python
#Calculate the correlation between the Ozone AQI Value and the NO2 AQI Value

correlation = data['Ozone AQI Value'].corr(data['NO2 AQI Value'])

plt.scatter(data['Ozone AQI Value'], data['NO2 AQI Value'], alpha=0.7)
plt.xlabel('Ozone AQI Value')
plt.ylabel('NO2 AQI Value')
plt.title('Correlation: Ozone AQI Value and NO2 AQI Value')
plt.show()
print("Correlation between Ozone AQI Value and NO2 AQI Value: {:.2f}".format(correlation))
```

```python
#Get the latitude and longitude of the city with the lowest CO AQI Value.

city_lowest_co_aqi = data.loc[data['CO AQI Value'].idxmin(), 'City']
lowest_co_aqi_lat = data.loc[data['CO AQI Value'].idxmin(), 'lat']
lowest_co_aqi_lng = data.loc[data['CO AQI Value'].idxmin(), 'lng']

print("City with the lowest CO AQI Value: {}".format(city_lowest_co_aqi))
print("Latitude: {}".format(lowest_co_aqi_lat))
print("Longitude: {}".format(lowest_co_aqi_lng))
```

```python
#Count the number of rows where the PM2.5 AQI Category is "Moderate" and the NO2 AQI Category is "Unhealthy".

filtered_data = data[(data['PM2.5 AQI Category'] == 'Unhealthy') & (data['NO2 AQI Category'] == 'Good')]
count = len(filtered_data)
print(' number of rows where the PM2.5 AQI Category is "Moderate" and the NO2 AQI Category is "Unhealthy":', count)
```

```python
#Get the top 5 countries with the highest average CO AQI Value

average_co_aqi_by_country = data.groupby('Country')['CO AQI Value'].mean()
top_countries = average_co_aqi_by_country.nlargest(5)

plt.bar(top_countries.index, top_countries.values)
plt.xlabel('Country')
plt.ylabel('Average CO AQI Value')
plt.title('Top 5 Countries with Highest Average CO AQI Value')
plt.xticks(rotation=45)
plt.show()
print(top_countries)
```

```python
# Filter rows where AQI Category is "Hazardous" and PM2.5 AQI Category is "Very Unhealthy"

filtered_data = data[(data['AQI Category'] == 'Hazardous') & (data['PM2.5 AQI Category'] == 'Very Unhealthy')]

plt.scatter(filtered_data['lat'], filtered_data['lng'])
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Locations with AQI Category "Hazardous" and PM2.5 AQI Category "Very Unhealthy"')
plt.show()
print(filtered_data)
```
